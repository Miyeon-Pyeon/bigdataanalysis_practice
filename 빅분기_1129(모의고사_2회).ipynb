## 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
df.head()
```

```python
# 성별이 Male인 환자들의 age의 평균값은 ?

df['age'] = df['age'].str.replace('*','').astype('int')
df[df['gender'] == 'Male'].age.mean()
```

```python
# bmi컬럼의 결측치를 bmi컬럼의 결측치를 제외한 나머지 값들의 중앙값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

df['bmi'] = df['bmi'].fillna(df['bmi'].median())
round(df['bmi'].mean(), 3)
```

```python
# bmi컬럼의 각 결측치들을 직전의 행의 bmi값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

df['bmi'] = df['bmi'].fillna(method='ffill')
round(df['bmi'].mean(), 3)
```

```python
# bmi컬럼의 각 결측치들을 결측치를 가진 환자 나이대(10단위)의 평균 bmi 값으로 대체한 후 대체된 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

mean = df[df.bmi.notnull()].groupby(df.age//10*10).bmi.mean()
dic = {x:y for x,y in mean.items()}

idx = df.loc[df.bmi.isnull(),['age','bmi']].index

df.loc[df.bmi.isnull(),'bmi'] = (df[df.bmi.isnull()].age//10*10).map(lambda x : dic[x])

df.bmi.mean()
```

```python
# avg_glucose_level 컬럼의 값이 200이상인 데이터를 모두 199로 변경하고 stroke값이 1인 데이터의 avg_glucose_level값의 평균을 소수점이하 3자리 까지 구하여라

df.loc[df.avg_glucose_level >= 200, 'avg_glucose_level'] = 199
round(df[df.stroke == 1].avg_glucose_level.mean(), 3)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

```python
# Attack컬럼의 값을 기준으로 내림차순정렬 했을때 상위 400위까지 포켓몬들과 401~800위까지의 포켓몬들에서 전설포켓몬(Legendary컬럼)의 숫자 차이는?

high = df.sort_values(by='Attack', ascending=False).iloc[:400,:]
low = df.sort_values(by='Attack', ascending=False).iloc[400:,:]

high_l = high['Legendary'].sum()
low_l = low['Legendary'].sum()

print(high_l - low_l)
```

```python
# Type 1 컬럼의 종류에 따른 Total 컬럼의 평균값을 내림차순 정렬했을때 상위 3번째 Type 1은 무엇인가?

df.groupby('Type 1')['Total'].mean().sort_values(ascending=False).index[2]
```

```python
# 결측치가 존재하는 행을 모두 지운 후 처음부터 순서대로 60% 데이터를 추출하여 Defense컬럼의 1분위수를 구하여라

df.dropna()[:int(len(df.dropna()) * 0.6)].Defense.quantile(0.25)
```

```python
# Type 1 컬럼의 속성이 Fire인 포켓몬들의 Attack의 평균이상인 Water속성의 포켓몬 수를 구하여라

# df[df['Type 1'] == 'Water'] >= df[df['Type 1'] == 'Fire'].Attack.mean()   >> 오답

target = df[df.Attack >= df[df['Type 1'] == 'Fire'].Attack.mean()]
target[target['Type 1'] == 'Water'].shape[0]
```

```python
# 각 세대 중(Generation 컬럼)의 Speed와 Defense 컬럼의 차이(절댓값)이 가장 큰 세대는?

#df.groupby('Generation')['Speed'] - df.groupby('Generation')['Defense']  >> 오답

abs(df.groupby(['Generation'])[['Speed','Defense']].mean().T.diff().T).sort_values('Defense').index[-1]
```

## 작업 2유형

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')

display(train.head())
display(test.head())
```

```python
print(train.info())  # gender, age, ever_married, work_type, Residence_type, smoking_status
print(test.info())
```

```python
X_train = train.drop(columns = ['id','stroke'])
y = train['stroke']
X_test = test.drop(columns=['id'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_col = X_train.select_dtypes(exclude='object').columns

X_train[num_col] = ss.fit_transform(X_train[num_col])
X_test[num_col] = ss.transform(X_test[num_col])
```

```python
print(set(X_test['gender']) - set(X_train['gender']))
print(set(X_test['age']) - set(X_train['age']))
print(set(X_test['ever_married']) - set(X_train['ever_married']))
print(set(X_test['work_type']) - set(X_train['work_type']))
print(set(X_test['Residence_type']) - set(X_train['Residence_type']))
print(set(X_test['smoking_status']) - set(X_train['smoking_status']))
```

```python
import numpy as np

X_test['age'] = X_test['age'].apply(lambda x: x if x in X_train['age'].unique() else np.nan)

# NaN 처리 (예: 최빈값으로 대체)
X_test['age'] = X_test['age'].fillna(X_train['age'].mode()[0])

print(set(X_test['age']) - set(X_train['age']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['gender'] = le.fit_transform(X_train['gender'])
X_test['gender'] = le.transform(X_test['gender'])

X_train['age'] = le.fit_transform(X_train['age'])
X_test['age'] = le.transform(X_test['age'])

X_train['ever_married'] = le.fit_transform(X_train['ever_married'])
X_test['ever_married'] = le.transform(X_test['ever_married'])

X_train['work_type'] = le.fit_transform(X_train['work_type'])
X_test['work_type'] = le.transform(X_test['work_type'])

X_train['Residence_type'] = le.fit_transform(X_train['Residence_type'])
X_test['Residence_type'] = le.transform(X_test['Residence_type'])

X_train['smoking_status'] = le.fit_transform(X_train['smoking_status'])
X_test['smoking_status'] = le.transform(X_test['smoking_status'])
```

```python
print(X_train.info())
print(X_test.info())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_val, pred)
print(roc)
```

```python
pred2 = model.predict_proba(X_test)[:,1]
```

```python
submission = pd.DataFrame()
submission['id'] = test['id']
submission['proba'] = pred2
submission.head()
```

```python
submission.to_csv('stroke.csv', index=False)
```
