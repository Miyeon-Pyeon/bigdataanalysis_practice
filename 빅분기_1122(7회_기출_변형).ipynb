## 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p1.csv')
df.head(5)
```

```python
# 국어,수학,영어,과학 과목 중 가장 많은 학생들이 응시한 시험을 선택하고 해당과목의 점수를 표준화 했을 때 가장 큰 표준화 점수를 구하여라

import numpy as np
sub = df.iloc[:,1:].count().idxmax()
std = df[sub]

r = ((df[sub] - df[sub].mean()) / df[sub].std(ddof=0))
print(r)

from scipy.stats import zscore
result = zscore(df[sub].dropna()).max()
print(result)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3.csv')
df.head(5)
```

```python
# var_6 컬럼의 1,3사분위수 각각 IQR의 1.5배 벗어난 이상치의 숫자를 구하라

one = df['var_6'].quantile(0.25)
three = df['var_6'].quantile(0.75)
iqr = three - one

min_ = one - 1.5 * iqr
max_ = three + 1.5 * iqr

ans = df[(df['var_6'] < min_) | (df['var_6'] > max_)].shape[0]
print(ans)
```

## 작업 2유형

```python
# 종속변수 :이용금액 , 평가지표 : rmse

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info())  # ID, 업종명, 이용자구분, 성별
print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

le_id = LabelEncoder()
le_industry = LabelEncoder()
le_user = LabelEncoder()
le_gender = LabelEncoder()

train['ID'] = le_id.fit_transform(train['ID'])
train['업종명'] = le_industry.fit_transform(train['업종명'])
train['이용자구분'] = le_user.fit_transform(train['이용자구분'])
train['성별'] = le_gender.fit_transform(train['성별'])

test['업종명'] = le_industry.transform(test['업종명'])
test['이용자구분'] = le_user.transform(test['이용자구분'])
test['성별'] = le_gender.transform(test['성별'])
```

```python
print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['이용금액']
X = train.drop(columns = ['이용금액','ID'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)  # 3.180065609147414e+16
# print(mse)

rmse = np.sqrt(mse)
print(rmse)
```

```python
test_X = test.drop(columns=['ID'])
pred2 = model.predict(test_X)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
```

```python
submission.to_csv('card.csv', index=False)
```

## 작업 3유형

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_1.csv')
df.head()
```

```python
# 선형관계 가장 큰 변수 찾아 상관계수를 구하여라

ans = df.corr()['Target'].sort_values()[-2]
print(ans)
```

```python
# Target 변수를 종속변수로 하여 다중선형회귀모델링을 진행했을 때 v2 컬럼의 회귀 계수는?

import numpy as np
import pandas as pd
import statsmodels.api as sm

X = df.drop('Target', axis=1)
X = sm.add_constant(X)
y = df['Target']

model = sm.OLS(y, X).fit()
print(model.params['v2'])
```

```python
# 회귀 계수들이 가지는 p값들 중 최대 값은?

ans = model.pvalues.max()
print(ans)
```
