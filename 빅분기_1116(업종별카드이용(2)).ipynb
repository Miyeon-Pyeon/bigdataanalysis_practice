```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info()) # 업종명, 이용자구분, 성별
print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['업종명'] = le.fit_transform(train['업종명'])
train['이용자구분'] = le.fit_transform(train['이용자구분'])
train['성별'] = le.fit_transform(train['성별'])

test['업종명'] = le.fit_transform(test['업종명'])
test['이용자구분'] = le.fit_transform(test['이용자구분'])
test['성별'] = le.fit_transform(test['성별'])
```

```python
# print(train.info())
# print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['이용금액']
X = train.drop(columns=['이용금액','ID'])

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 10)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)
# print('mse: ',mse)

rmse = np.sqrt(mse)
print('rmse: ', rmse)
```

```python
test_X_data = test.drop(columns=['ID'])
pred2 = model.predict(test_X_data)
```

```python
sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['price']  = pred2
sumission.head()
```

```python
sumission.to_csv('card.csv', index=False)
```
