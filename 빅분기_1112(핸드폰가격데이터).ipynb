```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")
df.head()
```

```python
# price_range 의 각 value를 그룹핑하여 각 그룹의 n_cores 의 빈도가 가장높은 value와 그 빈도수를 구하여라

# df.groupby('price_range', as_index=False).n_cores.value_counts().iloc[0] > 오답

df[['price_range','n_cores']].groupby(['price_range','n_cores']).size().sort_values().groupby(level=0).tail(1)
```

```python
# price_range 값이 3인 그룹에서 상관관계가 2번째로 높은 두 컬럼과 그 상관계수를 구하여라

c = df[df['price_range'] == 3].corr().unstack().sort_values(ascending=False)
ans = c.loc[c != 1].reset_index().iloc[1]
ans
```
