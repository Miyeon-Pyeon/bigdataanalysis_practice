# 회귀

```python
# 종속변수 :이용금액 , 평가지표 : rmse

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info()) # 업종명, 이용자구분, 성별
print(test.info())
```

```python
X_train = train.drop(columns=['ID','이용금액'])
y = train['이용금액']
X_test = test.drop(columns=['ID'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_col = X_train.select_dtypes(exclude='object').columns

X_train[num_col] = ss.fit_transform(X_train[num_col])
X_test[num_col] = ss.transform(X_test[num_col])
```

```python
print(set(X_test['업종명']) - set(X_train['업종명']))
print(set(X_test['이용자구분']) - set(X_train['이용자구분']))
print(set(X_test['성별']) - set(X_train['성별']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['업종명'] = le.fit_transform(X_train['업종명'])
X_test['업종명'] = le.transform(X_test['업종명'])

X_train['이용자구분'] = le.fit_transform(X_train['이용자구분'])
X_test['이용자구분'] = le.transform(X_test['이용자구분'])

X_train['성별'] = le.fit_transform(X_train['성별'])
X_test['성별'] = le.transform(X_test['성별'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
import numpy as np

mse = mean_squared_error(y_val, pred)  # 6.12607596893885e+16
print('mse: ',mse)
rmse = np.sqrt(mse)
print('rmse: ',rmse)
mae = mean_absolute_error(y_val,pred)
print('mae: ', mae)
mape = mean_absolute_percentage_error(y_val, pred)
print('mape: ', mape)
r2 = r2_score(y_val, pred)
print('r2_score: ', r2)
```

```python
pred2 = model.predict(X_test)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
```

```python
submission.to_csv('jeju.csv', index=False)
```

```python
# 예측 변수 General_Health, test.csv에 대해 ID별로 General_Health 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 General_Health 두개만 존재해야함. 평가지표는 f1score

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info())  # Checkup, Exercise, Heart_Disease, Skin_Cancer, Other_Cancer, Depression, Diabetes, Arthritis, Sex, Age_Category, Smoking_History
print(test.info())
```

```python
X_train = train.drop(columns = ['ID','General_Health'])
y = train['General_Health']
X_test = test.drop(columns = ['ID'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_col = X_train.select_dtypes(exclude='object').columns

X_train[num_col] = ss.fit_transform(X_train[num_col])
X_test[num_col] = ss.transform(X_test[num_col])
```

```python
print(set(X_test['Checkup']) - set(X_train['Checkup']))
print(set(X_test['Exercise']) - set(X_train['Exercise']))
print(set(X_test['Heart_Disease']) - set(X_train['Heart_Disease']))
print(set(X_test['Skin_Cancer']) - set(X_train['Skin_Cancer']))
print(set(X_test['Other_Cancer']) - set(X_train['Other_Cancer']))
print(set(X_test['Depression']) - set(X_train['Depression']))
print(set(X_test['Diabetes']) - set(X_train['Diabetes']))
print(set(X_test['Arthritis']) - set(X_train['Arthritis']))
print(set(X_test['Sex']) - set(X_train['Sex']))
print(set(X_test['Age_Category']) - set(X_train['Age_Category']))
print(set(X_test['Smoking_History']) - set(X_train['Smoking_History']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['Checkup'] = le.fit_transform(X_train['Checkup'])
X_test['Checkup'] = le.transform(X_test['Checkup'])

X_train['Exercise'] = le.fit_transform(X_train['Exercise'])
X_test['Exercise'] = le.transform(X_test['Exercise'])

X_train['Heart_Disease'] = le.fit_transform(X_train['Heart_Disease'])
X_test['Heart_Disease'] = le.transform(X_test['Heart_Disease'])

X_train['Skin_Cancer'] = le.fit_transform(X_train['Skin_Cancer'])
X_test['Skin_Cancer'] = le.transform(X_test['Skin_Cancer'])

X_train['Other_Cancer'] = le.fit_transform(X_train['Other_Cancer'])
X_test['Other_Cancer'] = le.transform(X_test['Other_Cancer'])

X_train['Depression'] = le.fit_transform(X_train['Depression'])
X_test['Depression'] = le.transform(X_test['Depression'])

X_train['Diabetes'] = le.fit_transform(X_train['Diabetes'])
X_test['Diabetes'] = le.transform(X_test['Diabetes'])

X_train['Arthritis'] = le.fit_transform(X_train['Arthritis'])
X_test['Arthritis'] = le.transform(X_test['Arthritis'])

X_train['Sex'] = le.fit_transform(X_train['Sex'])
X_test['Sex'] = le.transform(X_test['Sex'])

X_train['Age_Category'] = le.fit_transform(X_train['Age_Category'])
X_test['Age_Category'] = le.transform(X_test['Age_Category'])

X_train['Smoking_History'] = le.fit_transform(X_train['Smoking_History'])
X_test['Smoking_History'] = le.transform(X_test['Smoking_History'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_val)
#pred_p = model.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score, recall_score, precision_score

f1 = f1_score(y_val, pred, average='macro')
print('f1: ', f1)

# roc = roc_auc_score(y_val, pred_p)
# print('roc: ', roc)

accuracy = accuracy_score(y_val, pred)
print('accuracy: ', accuracy)

recall = recall_score(y_val, pred, average='macro')
print('recall: ', recall)

precision = precision_score(y_val, pred, average='macro')
print('precision: ', precision)
```

```python
pred2 = model.predict(X_test)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
```

```python
submission.to_csv('GH.csv', index=False)
```
