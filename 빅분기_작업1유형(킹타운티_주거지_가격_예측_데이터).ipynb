```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice//train.csv")
df.head()
```

```python
#95
# bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라

Ans = df[df['bedrooms'] == df['bedrooms'].value_counts().index[0]].price.quantile(0.9) - df[df['bedrooms'] == df['bedrooms'].value_counts().index[0]].price.quantile(0.1)
print(Ans)
```
