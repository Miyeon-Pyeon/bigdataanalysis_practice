```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/audit/train.csv")
df.head()
```

```python
#99
# 데이터의 Risk 값에 따른 score_a와 score_b의 평균값을 구하여라

Ans = df.groupby('Risk')[['Score_A','Score_B']].mean()
print(Ans)
```
