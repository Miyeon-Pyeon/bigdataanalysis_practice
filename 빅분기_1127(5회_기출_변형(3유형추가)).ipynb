```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
df.head(5)
```

```python
# 20L가격과 5L가격이 모두 0원이 아닌 데이터만 필터를 한 후, 각 row별로 20L가격과 5L가격의 차이를 ‘차이가격’ 이라 부른다고 하자. 시도명 별 차이가격의 평균가격을 비교할때 그 값이 가장 큰 금액을 반올림하여 소숫점 이하 1자리까지 구하여라

f_df = df[(df['20L가격'] != 0) & (df['5L가격'] != 0)]
f_df['차이가격'] = f_df['20L가격'] - f_df['5L가격']
ans = f_df.groupby('시도명')['차이가격'].mean().sort_values(ascending=False).values[0]
print(round(ans,1))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)
```

```python
# BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 고도 비반은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상 저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?

df['BMI'] = df['weight(kg)'] / ((df['height(cm)']/100) * (df['height(cm)']/100))

fat = df[(df['BMI'] >= 25)].count()
slim = df[df['BMI'] < 18.5].count()

print(fat)
print(slim)

fat + slim
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_3.csv')
df.head(5)
```

```python
# 순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

df['순유입인원'] = df[[x for x in df.columns if '전입' in x]].sum(axis=1) - df[[x for x in df.columns if '전출' in x]].sum(axis=1)

df.groupby('년도')['순유입인원'].max().sum()
```

## 작업 2유형

```python
# 예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 price 두개만 존재해야함. 평가지표는 rmse

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info())  # model, transmission, fuelType
print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

# 모든 열을 문자열로 변환하여 데이터 타입 통일
train['model'] = train['model'].astype(str)
test['model'] = test['model'].astype(str)

train['transmission'] = train['transmission'].astype(str)
test['transmission'] = test['transmission'].astype(str)

train['fuelType'] = train['fuelType'].astype(str)
test['fuelType'] = test['fuelType'].astype(str)

# LabelEncoder 초기화
le_m = LabelEncoder()
le_t = LabelEncoder()
le_f = LabelEncoder()

# 각 열에 대해 train과 test 결합 후 fit
le_m.fit(pd.concat([train['model'], test['model']], axis=0))
le_t.fit(pd.concat([train['transmission'], test['transmission']], axis=0))
le_f.fit(pd.concat([train['fuelType'], test['fuelType']], axis=0))

# train 데이터 변환
train['model'] = le_m.transform(train['model'])
train['transmission'] = le_t.transform(train['transmission'])
train['fuelType'] = le_f.transform(train['fuelType'])

# test 데이터 변환
test['model'] = le_m.transform(test['model'])
test['transmission'] = le_t.transform(test['transmission'])
test['fuelType'] = le_f.transform(test['fuelType'])
```

```python
from sklearn.model_selection import train_test_split

y = train['price']
X = train.drop(columns=['ID','price'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)
print('mse : ', mse)
rmse = np.sqrt(mse)
print('rmse : ', rmse)
```

```python
test_X = test.drop(columns = ['ID'])
pred2 = model.predict(test_X)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
```

```python
submission.to_csv('price.csv', index=False)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

```python
# 55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지

import numpy as np
mean = np.round(np.mean(df.height), 3)
print(mean)
```

```python
# t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)

import numpy as np
from scipy.stats import t

std = np.std(df.height, ddof=1)
n = len(df.height)

confidence_level = 0.95
ddof = n -1

t_value = round(t.ppf((1+confidence_level) / 2, ddof), 4)
print(t_value)
```

```python
# 95% 신뢰구간을 구하여라(print(lower,upper) 방식으로 출력, 각각의 값은 소숫점 이하 3째자리까지)

lower = round(mean - t_value * std / np.sqrt(n), 3)
upper = round(mean + t_value * std / np.sqrt(n), 3)
print(lower, upper)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_2.csv')
df.head()
```

```python
# 3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 검정 통계량을 반올림하여 소숫점 이하 3자리까지 구하여라

from scipy.stats import kruskal
a = df[df['ID'] == 'A'].value.values
b = df[df['ID'] == 'B'].value.values
c = df[df['ID'] == 'C'].value.values

s, p = kruskal(a,b,c)

round_s = round(s,3)
print(round_s)
```

```python
# 3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 p-value를 반올림하여 소숫점 이하 3자리까지 구하여라. 귀무가설과 대립가설중 0.05 유의수준에서 유의한 가설을 출력하라

round_p = round(p, 3)
print(round_p)
```
