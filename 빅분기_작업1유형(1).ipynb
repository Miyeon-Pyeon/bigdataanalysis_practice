```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
#1
# 인기동영상 제작횟수가 많은 채널 상위 10개 출력(채널ID의 빈도수 중 상위 10개를 뽑아 리스트로 출력)

Ans = list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
print(Ans)
```

```python
#2
#논란으로 인기동영상이 된 케이스 확인(dislike가 like보다 많은 동영상)

Ans = list(df.loc[df['dislikes']>df['likes']].channelTitle.unique())
print(Ans)
```

```python
#3
#채널명을 바꾼 케이스 확인(channelId는 고유값임)(중복되는 )

change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
target = change[change>1]
print(len(target))
```

```python
#4
#일요일에 인기있었던 영상 중 가장 많은 영상의 종류

df ['trending_date2'] = pd.to_datetime(df['trending_date2'])
Ans = df.loc[df['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts().index[0]
print(Ans)
```
