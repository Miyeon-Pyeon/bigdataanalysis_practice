```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info()) # labelencoding :General_Health, Checkup, Exercise, Heart_Disease, Skin_Cancer, Other_Cancer, Depression, Diabetes, Arthritis, Sex, Age_Category, Smoking_History

print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

l = LabelEncoder()

train['Checkup'] = l.fit_transform(train['Checkup'])
train['Exercise'] = l.fit_transform(train['Exercise'])
train['Heart_Disease'] = l.fit_transform(train['Heart_Disease'])
train['Skin_Cancer'] = l.fit_transform(train['Skin_Cancer'])
train['Other_Cancer'] = l.fit_transform(train['Other_Cancer'])
train['Depression'] = l.fit_transform(train['Depression'])
train['Diabetes'] = l.fit_transform(train['Diabetes'])
train['Arthritis'] = l.fit_transform(train['Arthritis'])
train['Sex'] = l.fit_transform(train['Sex'])
train['Age_Category'] = l.fit_transform(train['Age_Category'])
train['Smoking_History'] = l.fit_transform(train['Smoking_History'])

test['Checkup'] = l.fit_transform(test['Checkup'])
test['Exercise'] = l.fit_transform(test['Exercise'])
test['Heart_Disease'] = l.fit_transform(test['Heart_Disease'])
test['Skin_Cancer'] = l.fit_transform(test['Skin_Cancer'])
test['Other_Cancer'] = l.fit_transform(test['Other_Cancer'])
test['Depression'] = l.fit_transform(test['Depression'])
test['Diabetes'] = l.fit_transform(test['Diabetes'])
test['Arthritis'] = l.fit_transform(test['Arthritis'])
test['Sex'] = l.fit_transform(test['Sex'])
test['Age_Category'] = l.fit_transform(test['Age_Category'])
test['Smoking_History'] = l.fit_transform(test['Smoking_History'])
```

```python
print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['General_Health']
X = train.drop(columns = ['ID','General_Health'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_test, pred, average='macro')

print(f1_score)
```

```python
test_X = test.drop(columns = ['ID'])
pred2 = model.predict(test_X)
```

```python
sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['price'] = pred2
sumission.head()
```

```python
sumission.to_csv('health.csv',index=False)
```
