```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nflx/NFLX.csv")
df.head()
```

```python
# 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라

df['Date'] = pd.to_datetime(df.Date)

# df.groupby(df['Date'].dt.year).[df['Date'].dt.month == 5].Open.mean()  > 오답

tar = df.groupby(df['Date'].dt.strftime('%Y-%m')).mean()
tar.loc[tar.index.str.contains('-05')].Open
```
