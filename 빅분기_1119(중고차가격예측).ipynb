```python
import pandas as pd

paths = ['audi.csv','bmw.csv','cclass.csv','focus.csv','ford.csv','hyundi.csv','merc.csv','skoda.csv','toyota.csv','vauxhall.csv','vw.csv']

merged_df = pd.DataFrame()

for file in paths:
    temp_df = pd.read_csv(file)  # 파일 읽기
    df = pd.concat([merged_df, temp_df], ignore_index=True)

print("병합된 데이터프레임 크기:", merged_df.shape)
df.head()
```

```python
df.info() # model, transmission, fuelType
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['model'] = le.fit_transform(df['model'])
df['transmission'] = le.fit_transform(df['transmission'])
df['fuelType'] = le.fit_transform(df['fuelType'])
```

```python
df.info()
```

```python
train = df.iloc[:12126, :]
test = df.iloc[12126:, :]

print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['price']
X = train.drop(columns=['price'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)  # 2244105.7044085665
# print(mse)
rmse = np.sqrt(mse)
print(rmse)
```

```python
test_X = test.drop(columns=['price'])
pred2 = model.predict(test_X)
print(pred2)
```

```python
submission = pd.DataFrame()
submission['price'] = pred2
submission.head()
```

```python
submission.to_csv('car_price.csv', index=False)
```
