```python
import pandas as pd
```

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/Jeju.csv'
df = pd.read_csv(DataUrl,encoding='euc-kr')
```

```python
df
```

```python
# 데이터 마지막 3개행을 출력하라

df.tail(3)
```

```python
# 수치형 변수를 가진 컬럼을 출력하라

# df.columns[int]  > 오답

df.select_dtypes(exclude=object).columns
```

```python
# 범주형 변수를 가진 컬럼을 출력하라

df.select_dtypes(include=object).columns
```

```python
# 각 컬럼의 결측치 숫자를 파악해라

df.isnull().sum()
```

```python
# 각 컬럼의 데이터 수, 데이터 타입을 한번에 확인하라

df.info()
```

```python
# 각 수치형 변수의 분포(사분위, 평균, 표준편차, 최대, 최소)를 확인하라

# df.quantiles()  > 오답

df.describe()
```

```python
# 거주인구 컬럼의 값들을 출력하라

df['거주인구']
```

```python
# 평균 속도 컬럼의 4분위 범위(IQR)값을 구하여라

# df['평균 속도'].quantiles()  > 오답

df['평균 속도'].quantile(0.75) - df['평균 속도'].quantile(0.25)
```

```python
# 읍면동명 컬럼의 유일값 갯수를 출력하라
# 기억하기

df['읍면동명'].nunique()
```

```python
# 읍면동명 컬럼의 유일값을 모두 출력하라
# 기억하기

df['읍면동명'].unique()
```
