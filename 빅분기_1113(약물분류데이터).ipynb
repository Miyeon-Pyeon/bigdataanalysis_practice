```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv")
df.head()
```

```python
# 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

df['Age2'] = (df['Age'] // 10) * 10

m = df[df['Sex'] == 'M']
m.groupby('Age2').Na_to_K.mean()
```
