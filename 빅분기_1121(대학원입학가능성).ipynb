```python
import pandas as pd

df = pd.read_csv('Admission_Predict.csv')
df.head()
```

```python
df.info()
```

```python
train = df.iloc[:320, :]
test = df.iloc[320:,:]

print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['Chance of Admit ']
X = train.drop(columns=['Chance of Admit ','Serial No.'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)  # 0.004000686093750002
#print(mse)
rmse = np.sqrt(mse)
print(rmse)
```

```python
test_X = test.drop(columns = ['Chance of Admit ','Serial No.'])
pred2 = model.predict(test_X)
print(pred2)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['Serial No.']
submission['Chance of Admit'] = pred2
submission.head()
```

```python
submission.to_csv('Admit.csv', index=False)
```
