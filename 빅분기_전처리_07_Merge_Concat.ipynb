## 07_Merge, Concat

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv')
df.head()
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
#91

plus = pd.concat([df1, df2])
Ans = plus
Ans.head()
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
#92
#        

hab = pd.concat([df3, df4], join='inner')
Ans = hab
Ans.head()
```

```python
#93

hab = pd.concat([df3, df4], join='outer').fillna(0)
Ans = hab
Ans.head()
```

```python
df5 = df.T.iloc[:8,:3]
df6 = df.T.iloc[7:,2:5]

df5.columns = ['Afghanistan', 'Albania', 'Algeria']
df6.columns = ['Afghanistan', 'Albania', 'Algeria']

display(df5)
display(df6)
```

```python
#94

Ans = pd.merge(df5, df6, on='Afghanistan', how='inner')
Ans
```

```python
#95

Ans =pd.merge(df5,df6,on='Algeria',how='outer')
Ans
```
