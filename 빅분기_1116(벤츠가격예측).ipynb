```python
# 예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 price 두개만 존재해야함. 평가지표는 rmse

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info()) # model, transmission, fuelType
print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['model'] = le.fit_transform(train['model'])
train['transmission'] = le.fit_transform(train['transmission'])
train['fuelType'] = le.fit_transform(train['fuelType'])

test['model'] = le.fit_transform(test['model'])
test['transmission'] = le.fit_transform(test['transmission'])
test['fuelType'] = le.fit_transform(test['fuelType'])
```

```python
# print(train.info())
# print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['price']
X = train.drop(columns = ['price','ID'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)
# print(mse)

rmse = np.sqrt(mse)
print('rmse: ', rmse)
```

```python
test_X_data = test.drop(columns = ['ID'])
pred2 = model.predict(test_X_data)
# print(pred2)
```

```python
sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['price'] = pred2
sumission.head()
```

```python
sumission.to_csv('benz.csv', index=False)
```
