## 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_1.csv')
df.head(5)
```

```python
# Temperature컬럼에서 숫자가 아닌 문자들을 제거후 숫자 타입으로 바꾸고 3분위수에서 1분위수의 차이를 소숫점 이하 2자리까지 구하여라

df['Temperature_f'] = df['Temperature'].str.extract(r'([\d\.]+)').astype('float')
ans = df['Temperature_f'].quantile(0.75) - df['Temperature_f'].quantile(0.25)
print(round(ans, 2))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_2.csv')
df.head(5)
```

```python
# Likes를 Comments로 나눈 비율이 20이상이면서 Keyword값이 minecraft인 영상들의 Views값의 평균을 정수로 구하여라

df['ratio'] = df['Likes'] / df['Comments']
ans = df[(df['ratio'] >= 20) & (df['Keyword'] == 'minecraft')]['Views'].mean()

print(int(ans))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_3.csv')
df.head(5)
```

```python
# date_added가 2018년 1월 이면서 country가 United Kingdom 단독 제작인 데이터의 갯수

df['date_added'] = pd.to_datetime(df['date_added'])
result = df[(df.date_added.dt.strftime("%Y-%m") =='2018-01') & (df.country =='United Kingdom')].shape[0]
print(result)
```

## 작업 2유형

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info())  # 결측치: Ever_Married, Graduated, Profession, Work_Experience, Family_Size, Var_1
print(test.info())  # 수치화: Gender, Ever_Married, Graduated, Profession, Spending_Score, Var_1, Segmentation
```

```python
X_train = train.drop(columns = ['ID','Segmentation'])
y = train['Segmentation']
X_test = test.drop(columns=['ID'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
X_train['Ever_Married'] = X_train['Ever_Married'].fillna('unknown')
X_train['Graduated'] = X_train['Graduated'].fillna('unknown')
X_train['Profession'] = X_train['Profession'].fillna('unknown')
X_train['Work_Experience'] = X_train['Work_Experience'].fillna(0)
X_train['Family_Size'] = X_train['Family_Size'].fillna(0)
X_train['Var_1'] = X_train['Var_1'].fillna('unknown')

X_test['Ever_Married'] = X_test['Ever_Married'].fillna('unknown')
X_test['Graduated'] = X_test['Graduated'].fillna('unknown')
X_test['Profession'] = X_test['Profession'].fillna('unknown')
X_test['Work_Experience'] = X_test['Work_Experience'].fillna(0)
X_test['Family_Size'] = X_test['Family_Size'].fillna(0)
X_test['Var_1'] = X_test['Var_1'].fillna('unknown')

# print(X_train.isna().sum())
# print(X_test.isna().sum())
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_columns = X_train.select_dtypes(exclude='object').columns
# print(num_columns)

X_train[num_columns] = ss.fit_transform(X_train[num_columns])
X_test[num_columns] = ss.transform(X_test[num_columns])
```

```python
print(set(X_test['Gender']) - set(X_train['Gender']))
print(set(X_test['Ever_Married']) - set(X_train['Ever_Married']))
print(set(X_test['Graduated']) - set(X_train['Graduated']))
print(set(X_test['Profession']) - set(X_train['Profession']))
print(set(X_test['Spending_Score']) - set(X_train['Spending_Score']))
print(set(X_test['Var_1']) - set(X_train['Var_1']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['Gender'] = le.fit_transform(X_train['Gender'])
X_test['Gender'] = le.transform(X_test['Gender'])
X_train['Ever_Married'] = le.fit_transform(X_train['Ever_Married'])
X_test['Ever_Married'] = le.transform(X_test['Ever_Married'])
X_train['Graduated'] = le.fit_transform(X_train['Graduated'])
X_test['Graduated'] = le.transform(X_test['Graduated'])
X_train['Profession'] = le.fit_transform(X_train['Profession'])
X_test['Profession'] = le.transform(X_test['Profession'])
X_train['Spending_Score'] = le.fit_transform(X_train['Spending_Score'])
X_test['Spending_Score'] = le.transform(X_test['Spending_Score'])
X_train['Var_1'] = le.fit_transform(X_train['Var_1'])
X_test['Var_1'] = le.transform(X_test['Var_1'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)

# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_val)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_val, pred, average='macro')
print(f1)
```

```python
pred2 = model.predict(X_test)
print(pred2)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['Segmentation'] = pred2
submission.head()
```

```python
submission.to_csv('Segmentation.csv', index=False)
```
