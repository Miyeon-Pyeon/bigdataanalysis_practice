## 작업 2유형

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p2_test_.csv')
train.head()
```

```python
print(train.info())  # Employment Type, GraduateOrNot, FrequentFlyer, EverTravelledAbroad
print(test.info())
```

```python
X_train = train.drop(columns = ['ID','TravelInsurance'])
y = train['TravelInsurance']
X_test = test.drop(columns=['ID'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_columns = X_train.select_dtypes(exclude='object').columns
X_train[num_columns] = ss.fit_transform(X_train[num_columns])
X_test[num_columns] = ss.transform(X_test[num_columns])
```

```python
print(set(X_test['Employment Type']) - set(X_train['Employment Type']))
print(set(X_test['GraduateOrNot']) - set(X_train['GraduateOrNot']))
print(set(X_test['FrequentFlyer']) - set(X_train['FrequentFlyer']))
print(set(X_test['EverTravelledAbroad']) - set(X_train['EverTravelledAbroad']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['Employment Type'] = le.fit_transform(X_train['Employment Type'])
X_test['Employment Type'] = le.transform(X_test['Employment Type'])
X_train['GraduateOrNot'] = le.fit_transform(X_train['GraduateOrNot'])
X_test['GraduateOrNot'] = le.transform(X_test['GraduateOrNot'])
X_train['FrequentFlyer'] = le.fit_transform(X_train['FrequentFlyer'])
X_test['FrequentFlyer'] = le.transform(X_test['FrequentFlyer'])
X_train['EverTravelledAbroad'] = le.fit_transform(X_train['EverTravelledAbroad'])
X_test['EverTravelledAbroad'] = le.transform(X_test['EverTravelledAbroad'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size = 0.2)

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_val, pred)
print(roc)
```

```python
pred2 = model.predict_proba(X_test)[:,1]
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['proba'] = pred2
submission.head()
```

```python
submission.to_csv('trip.csv', index=False)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_1.csv')
df.head()
```

```python
# A등급에 해당하는 유저는 몇명인지 확인하라

df['gap'] = df['투약후'] - df['투약전']

def filte(x):
  if x <= -3:
    return 'A'
  elif x <= 0:
    return 'B'
  else:
    return 'C'

df['groupby'] = df['gap'].apply(filte)
df['groupby'].value_counts().sort_index()['A']
```
