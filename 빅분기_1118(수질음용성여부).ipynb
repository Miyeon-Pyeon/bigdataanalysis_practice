```python
import pandas as pd

df = pd.read_csv('water_potability.csv')
df.head()
```

```python
print(df.info())  # ph, Sulfate, Trihalomethanes
```

```python
# 결측치 처리

df['ph'].fillna(df['ph'].mean(), inplace = True)
df['Sulfate'].fillna(df['Sulfate'].mean(), inplace = True)
df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean(), inplace = True)
```

```python
df.info()
```

```python
train = df.iloc[:2621, :]
test = df.iloc[-655:, :]
```

```python
print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['Potability']
X = train.drop(columns = ['Potability'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_test, pred, average='macro')
print(f1_score)
```

```python
test_X = test.drop(columns= ['Potability'])
pred2 = model.predict(test_X)
```

```python
submission = pd.DataFrame()
submission['Potability'] = pred2
submission.head()
```

```python
submission.to_csv('water.csv', index=False)
```
