```python
import pandas as pd
df = pd.read_csv('drug200.csv')
df.head()
```

```python
df.info() # Sex, BP, Cholesterol, Drug
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])
df['BP'] = le.fit_transform(df['BP'])
df['Cholesterol'] = le.fit_transform(df['Cholesterol'])
df['Drug'] = le.fit_transform(df['Drug'])
```

```python
train = df.iloc[:160, :]
test = df.iloc[160:, :]

print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['Drug']
X = train.drop(columns=['Drug'])

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_test, pred, average='macro')
print(f1_score)
```

```python
test_X = test.drop(columns = ['Drug'])
pred2 = model.predict(test_X)
```

```python
submission = pd.DataFrame()
submission['Drug'] = pred2
submission.head()
```

```python
submission.to_csv('drug.csv', index=False)
```
