```python
import pandas as pd

df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()  # sex, smoker, region
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])
```

```python
train = df.iloc[:1070, :]
test = df.iloc[-268:, :]

print(train.info())
print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['charges']
X = train.drop(columns=['charges'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, pred)  # 24254464.15266382
# print(mse)
rmse = np.sqrt(mse)
print(rmse)
```

```python
test_X = test.drop(columns = ['charges'])
pred2 = model.predict(test_X)
#print(pred2)
```

```python
submission = pd.DataFrame()
submission['charges'] = pred2
submission.head()
```

```python
submission.to_csv('doctor.csv', index=False)
```
