```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
df.head()
```

```python
#94
# 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

high = df[df['smoker'] == 'yes'].charges.quantile(0.9)
high2 = df[df['smoker'] == 'no'].charges.quantile(0.9)
yes = df[(df['smoker'] == 'yes') & (df['charges'] >= high)].charges.mean()
no = df[(df['smoker'] == 'no') & (df['charges'] >= high2)].charges.mean()
Ans = yes - no
print(Ans)
```
