## 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/spotify/spotify.csv')
df.head()
```

```python
# 데이터는 현재 년도별 100곡이 인기순으로 정렬되어 있다. 각 년도별 1~100위의 랭킹을 나타내는 rank컬럼을 만들고 매년도 1위의 bpm컬럼의 평균값을 구하여라

df = df.dropna()
df['rank'] = list(range(1,101)) * 10
df[df['rank'] == 1].bpm.mean()
```

```python
# 2015년도에 가장많은 top100곡을 올린 artist는 누구인가?

df[df['top year'] == 2015.0].artist.value_counts().index[0]
```

```python
# 년도별 rank값이 1~10위 까지의 곡들 중 두번째로 많은 top genre는 무엇인가?

df[(df['rank'] >= 1) & (df['rank'] <= 10)]['top genre'].value_counts().index[1]
```

```python
# 피처링의 경우 title에 표시된다. 피처링을 가장 많이 해준 가수는 누구인가?

df.title.str.split('feat.').str[1].dropna().str[:-1].str.strip().value_counts().index[0]
```

```python
# top year 년도를 기준으로 발매일(year released)과 top100에 진입한 일자 (top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?

df_f = df[df['year released'] != df['top year']]

int(df_f['top year'].value_counts().index[0])
```

```python
# artist 컬럼의 값에 대소문자 상관없이 q 단어가 들어가는 아티스트는 몇명인가?

df[df['artist'].str.lower().str.contains('q')].artist.nunique()
```

```python
# 년도 상관없이 전체데이터에서 1~50위와 51~100위간의 dur 컬럼의 평균값의 차이는?

top = df[(df['rank']>=1) & (df['rank']<=50)].dur.mean()
low = df[(df['rank']>=51) & (df['rank']<=100)].dur.mean()

print(abs(top-low))
```

```python
# title을 띄어쓰기 단어로 구분 했을때 가장 많이 나온 단어는 무엇인가? (대소문자 구분 x)

df['title'].str.split('\(feat').str[0].str.split().explode().str.lower().value_counts().index[0]
```

```python
# 년도별 nrgy값의 평균값을 구할때 최대 평균값과 최소 평균값의 차이를 구하여라

big = df.groupby('top year').nrgy.mean().values[0]
small = df.groupby('top year').nrgy.mean().values[-1]

print(abs(big - small))
```

```python
# artist중 artist type 타입을 여러개 가지고 있는 artist는 누구인가

# df[df.groupby('artist')['artist type'].value_counts()] >> 오답

df[['artist','artist type']].value_counts().reset_index().artist.value_counts().index[0]
```

## 작업 2유형

```python
import pandas as pd
#데이터 로드
X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_train.csv")
y = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_train.csv")
test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_test.csv")


display(X_train.head())
display(y.head())
display(test.head())
```

```python
X_train  = X_train.drop(columns=['ID'])
y = y.drop(columns=['ID'])
X_test  = test.drop(columns=['ID'])
```

```python
print(X_train.info())
print(y.info())
print(X_test.info())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_val, pred)
print(roc)
```

```python
pred2 = model.predict_proba(X_test)[:,1]
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['pose'] = pred2
submission.head()
```

```python
submission.to_csv('poes.csv',index=False)
```
