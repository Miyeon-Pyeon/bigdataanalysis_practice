## 작업 2유형

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

display(train.head())
display(test.head())
display(submission.head())
```

```python
print(train.info())  # job, marital, education, default, housing, loan, contact, month, poutcome
print(test.info())
```

```python
X_train = train.drop(columns=['ID','y'])
y = train['y']
X_test = test.drop(columns=['ID'])

print(X_train.shape, y.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

num_columns = X_train.select_dtypes(exclude='object').columns

X_train[num_columns] = ss.fit_transform(X_train[num_columns])
X_test[num_columns] = ss.transform(X_test[num_columns])
```

```python
print(set(X_test['job']) - set(X_train['job']))
print(set(X_test['marital']) - set(X_train['marital']))
print(set(X_test['education']) - set(X_train['education']))
print(set(X_test['default']) - set(X_train['default']))
print(set(X_test['housing']) - set(X_train['housing']))
print(set(X_test['loan']) - set(X_train['loan']))
print(set(X_test['contact']) - set(X_train['contact']))
print(set(X_test['month']) - set(X_train['month']))
print(set(X_test['poutcome']) - set(X_train['poutcome']))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X_train['job'] = le.fit_transform(X_train['job'])
X_test['job'] = le.transform(X_test['job'])

X_train['marital'] = le.fit_transform(X_train['marital'])
X_test['marital'] = le.transform(X_test['marital'])

X_train['education'] = le.fit_transform(X_train['education'])
X_test['education'] = le.transform(X_test['education'])

X_train['default'] = le.fit_transform(X_train['default'])
X_test['default'] = le.transform(X_test['default'])

X_train['housing'] = le.fit_transform(X_train['housing'])
X_test['housing'] = le.transform(X_test['housing'])

X_train['loan'] = le.fit_transform(X_train['loan'])
X_test['loan'] = le.transform(X_test['loan'])

X_train['contact'] = le.fit_transform(X_train['contact'])
X_test['contact'] = le.transform(X_test['contact'])

X_train['month'] = le.fit_transform(X_train['month'])
X_test['month'] = le.transform(X_test['month'])


X_train['poutcome'] = le.fit_transform(X_train['poutcome'])
X_test['poutcome'] = le.transform(X_test['poutcome'])
```

```python
# print(X_train.info())
# print(X_test.info())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size = 0.2)

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_val, pred)
print(roc)
```

```python
pred2 = model.predict_proba(X_test)[:,1]
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['proba'] = pred2
submission.head()
```

```python
submission.to_csv('bank.csv', index=False)
```
