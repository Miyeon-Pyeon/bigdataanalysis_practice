```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

```python
#92
# Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

Ans = df[df['Arrival Delay in Minutes'].isnull()].groupby(['Class','satisfaction'], as_index=False).size().pivot(index='Class', columns = 'satisfaction')
result = Ans[Ans['size']['neutral or dissatisfied'] < Ans['size']['satisfied']]
result
```
