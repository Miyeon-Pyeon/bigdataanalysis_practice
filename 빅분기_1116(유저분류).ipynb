```python
# 예측 변수 Segmentation, test.csv에 대해 ID별로 Segmentation의 클래스를 예측해서 저장후 제출,
# 제출 데이터 컬럼은 ID와 Segmentation 두개만 존재해야함. 평가지표는 macro f1 score

import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
print(train.info())
print(test.info())

# 결측치: Ever_Married, Graduated, Profession, Work_Experience, Family_Size, Var_1
# 수치화: Gender, Ever_Married, Graduated, Profession, Spending_Score, Var_1, Segmentation
```

```python
# 수치화

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['Gender'] = le.fit_transform(train['Gender'])
train['Ever_Married'] = le.fit_transform(train['Ever_Married'])
train['Graduated'] = le.fit_transform(train['Graduated'])
train['Profession'] = le.fit_transform(train['Profession'])
train['Spending_Score'] = le.fit_transform(train['Spending_Score'])
train['Var_1'] = le.fit_transform(train['Var_1'])
train['Segmentation'] = le.fit_transform(train['Segmentation'])

test['Gender'] = le.fit_transform(test['Gender'])
test['Ever_Married'] = le.fit_transform(test['Ever_Married'])
test['Graduated'] = le.fit_transform(test['Graduated'])
test['Profession'] = le.fit_transform(test['Profession'])
test['Spending_Score'] = le.fit_transform(test['Spending_Score'])
test['Var_1'] = le.fit_transform(test['Var_1'])
```

```python
# 결측치 처리

# train['Ever_Married'] = train['Ever_Married'].fillna(0)
train['Work_Experience'] = train['Work_Experience'].fillna(train['Work_Experience'].mean())
train['Family_Size'] = train['Family_Size'].fillna(train['Family_Size'].mean())
# train.info()

test['Work_Experience'] = test['Work_Experience'].fillna(test['Work_Experience'].mean())
test['Family_Size'] = test['Family_Size'].fillna(test['Family_Size'].mean())
# test.info()
```

```python
# print(train.info())
# print(test.info())
```

```python
from sklearn.model_selection import train_test_split

y = train['Segmentation']
X = train.drop(columns = ['Segmentation','ID'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_test, pred, average='macro')
print(f1_score)
```

```python
test_X = test.drop(columns = ['ID'])
pred2 = model.predict(test_X)
#print(pred2)
```

```python
sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['Segmentation'] = pred2
sumission.head()
```

```python
sumission.to_csv('user.csv', index=False)
```
