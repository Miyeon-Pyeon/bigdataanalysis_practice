```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

```python
# Legendary 컬럼은 전설포켓몬 유무를 나타낸다.전설포켓몬과 그렇지 않은 포켓몬들의 HP평균의 차이를 구하여라

t = df.groupby('Legendary')['HP'].mean()
ans = t.values[1] - t.values[0]
ans
```

```python
# Type 1은 주속성 Type 2 는 부속성을 나타낸다. 가장 많은 부속성 종류는 무엇인가?

df['Type 2'].value_counts().index[0]
```

```python
# 가장 많은 Type 1 의 종의 평균 Attack 을 평균 Defense로 나눈값은?

df['Type 1'].value_counts().index[0]

at = df[df['Type 1'] == 'Water'].Attack.mean()
de = df[df['Type 1'] == 'Water'].Defense.mean()

print(at / de)
```

```python
# 포켓몬 세대(Generation) 중 가장많은 Legendary를 보유한 세대는 몇세대인가?

# df.groupby('Generation').Legendary.value_counts()

df[df.Legendary == True].Generation.value_counts().index[0]
```

```python
# ‘HP’, ‘Attack’, ‘Defense’, ‘Sp. Atk’, ‘Sp. Def’, ‘Speed’ 간의 상관 계수중 가장 절댓값이 큰 두 변수와 그 값을 구하여라

col = ['HP', 'Attack', 'Defense', 'Sp. Atk','Sp. Def','Speed']
df[col].corr()
```

```python
# 각 Generation의 Attack으로 오름차순 정렬시 상위 3개 데이터들(18개)의 Attack의 전체 평균을 구하여라

df.sort_values(['Generation','Attack']).groupby('Generation').head(3).Attack.mean()
```

```python
# 각 Generation의 Attack으로 내림차순 정렬시 상위 5개 데이터들(30개)의 Attack의 전체 평균을 구하여라

df.sort_values(['Generation','Attack']).groupby('Generation').tail(5).Attack.mean()

# df.sort_values(['Generation','Attack'], ascending = False).groupby('Generation').head(5).Attack.mean()  > 얘도 정답
```

```python
# 가장 흔하게 발견되는 (Type1 , Type2) 의 쌍은 무엇인가?

df[['Type 1', 'Type 2']].value_counts().index[0]
```

```python
# 한번씩만 존재하는 (Type1 , Type2)의 쌍의 갯수는 몇개인가?

t = df[['Type 1', 'Type 2']].value_counts()
print(len(t[t ==1]))
```

```python
# 한번씩만 존재하는 (Type1 , Type2)의 쌍을 각 세대(Generation)은 각각 몇개씩 가지고 있는가?

t = df[['Type 1', 'Type 2']].value_counts()
t2 = t[t ==1]

lst = []
for value in t2.reset_index().values:
  T1 = value[0]
  T2 = value[1]

  sp = df[(df['Type 1'] == T1) & (df['Type 2'] == T2)]
  lst.append(sp)

pd.concat(lst).reset_index(drop=True).Generation.value_counts().sort_index()
```
