```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/diabetes/train.csv")
df.head()
```

```python
# Outcome 값에 따른 각 그룹의 각 컬럼의 평균 차이를 구하여라

df.groupby('Outcome').mean().diff().iloc[1,:]
```
