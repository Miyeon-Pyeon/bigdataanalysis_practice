```python
# 학교 세부유형이 일반중학교인 학교들 중 일반중학교 숫자가 2번째로 많은 시도의 일반중학교 데이터만 필터하여 해당 시도의 교원 한명 당 맡은 학생수가 가장 많은 학교를 찾아서 해당 학교의 교원수를 출력하라

ilban = df[df['학교세부유형']=='일반중학교']['시도'].value_counts().sort_values(ascending=False).index[1]  # 경기

dd = df[(df['학교세부유형']=='일반중학교') & (df['시도'] == ilban)]
dd['ratio'] = dd['일반학급_학생수_계'] / dd['교원수_총계_계']
dd.sort_values(by='ratio', ascending=False)['교원수_총계_계'].head(1)
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 준비
le_id = LabelEncoder()
le_ck = LabelEncoder()
le_ex = LabelEncoder()
le_hd = LabelEncoder()
le_sc = LabelEncoder()
le_oc = LabelEncoder()
le_dp = LabelEncoder()
le_di = LabelEncoder()
le_ar = LabelEncoder()
le_sex = LabelEncoder()
le_ac = LabelEncoder()
le_sh = LabelEncoder()

# LabelEncoder 학습
le_ck.fit(pd.concat([train['Checkup'], test['Checkup']]).unique())
le_ex.fit(pd.concat([train['Exercise'], test['Exercise']]).unique())
le_hd.fit(pd.concat([train['Heart_Disease'], test['Heart_Disease']]).unique())
le_sc.fit(pd.concat([train['Skin_Cancer'], test['Skin_Cancer']]).unique())
le_oc.fit(pd.concat([train['Other_Cancer'], test['Other_Cancer']]).unique())
le_dp.fit(pd.concat([train['Depression'], test['Depression']]).unique())
le_di.fit(pd.concat([train['Diabetes'], test['Diabetes']]).unique())
le_ar.fit(pd.concat([train['Arthritis'], test['Arthritis']]).unique())
le_sex.fit(pd.concat([train['Sex'], test['Sex']]).unique())
le_ac.fit(pd.concat([train['Age_Category'], test['Age_Category']]).unique())
le_sh.fit(pd.concat([train['Smoking_History'], test['Smoking_History']]).unique())

# LabelEncoder 변환
train['ID'] = le_id.fit_transform(train['ID'])
train['Checkup'] = le_ck.transform(train['Checkup'])
train['Exercise'] = le_ex.transform(train['Exercise'])
train['Heart_Disease'] = le_hd.transform(train['Heart_Disease'])
train['Skin_Cancer'] = le_sc.transform(train['Skin_Cancer'])
train['Other_Cancer'] = le_oc.transform(train['Other_Cancer'])
train['Depression'] = le_dp.transform(train['Depression'])
train['Diabetes'] = le_di.transform(train['Diabetes'])
train['Arthritis'] = le_ar.transform(train['Arthritis'])
train['Sex'] = le_sex.transform(train['Sex'])
train['Age_Category'] = le_ac.transform(train['Age_Category'])
train['Smoking_History'] = le_sh.transform(train['Smoking_History'])

test['Checkup'] = le_ck.transform(test['Checkup'])
test['Exercise'] = le_ex.transform(test['Exercise'])
test['Heart_Disease'] = le_hd.transform(test['Heart_Disease'])
test['Skin_Cancer'] = le_sc.transform(test['Skin_Cancer'])
test['Other_Cancer'] = le_oc.transform(test['Other_Cancer'])
test['Depression'] = le_dp.transform(test['Depression'])
test['Diabetes'] = le_di.transform(test['Diabetes'])
test['Arthritis'] = le_ar.transform(test['Arthritis'])
test['Sex'] = le_sex.transform(test['Sex'])
test['Age_Category'] = le_ac.transform(test['Age_Category'])
test['Smoking_History'] = le_sh.transform(test['Smoking_History'])
```

```python
# A 도시의 남성 600명과 여성 550명이 있다. 남성들 중 흡연자 비율은 0.2이며 여성들 중 흡연자 비율은 0.26이다.
# 남성과 여성 간에 흡연 여부에 따른 인구 비율이 다른지 확인하고 싶다. 유의 수준 0.05하 귀무가설에 대해 기각 / 채택 여부와 p-value값을 각각 출력하라

import numpy as np
from scipy.stats import chi2_contingency

t_m = 600
t_f = 550

smoking_ratio_m = 0.2
smoking_ratio_f = 0.26

smoking_m = int(t_m * smoking_ratio_m)
non_sm_m = int(t_m - smoking_m)

smoking_f = int(t_f * smoking_ratio_f)
non_sm_f = int(t_f - smoking_f)

data = np.array([[smoking_m, non_sm_m], [smoking_f, non_sm_f]])

chi2_stat, p_val, dof, expected = chi2_contingency(data)
print(p_val)

if p_val < 0.05:
  print('기각')
else:
  print('채택')
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/28/p7.csv')
df.head()
```

```python
# age와 Cholesterol을 가지고 weight를 예측하는 선형 회귀 모델을 만들려고한다. age의 회귀 계수를 구하여라

import statsmodels.api as sm
X = sm.add_constant(df[['age','Cholesterol']])
model = sm.OLS(df['weight'], X)
ans = model.fit()

print(ans.summary())

print(ans.params['age'])
```

```python
# age가 고정일 때 Cholesterol와 weight가 선형관계에 있다는 가설을 유의수준 0.05하에 검정하라

H = ans.pvalues['Cholesterol']

if H < 0.05:
  print('선형관계에 있다.')
else:
  print('선향관계가 아니다.')
```

```python
# age가 55, Cholesterol가 72.6일때 위 모델을 기반으로 weight값을 예측하라.

pred = ans.predict([1, 55, 72.6])  # const , age,Cholesterol
print(pred[0])
```
