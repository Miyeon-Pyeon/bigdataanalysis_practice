```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(train.head())
print(test.head())
```

```python
print(train.info()) # Gender, Customer Type, Type of Travel, Class, satisfaction
print(test.info())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['Gender'] = le.fit_transform(train['Gender'])
train['Customer Type'] = le.fit_transform(train['Customer Type'])
train['Type of Travel'] = le.fit_transform(train['Type of Travel'])
train['Class'] = le.fit_transform(train['Class'])
train['satisfaction'] = le.fit_transform(train['satisfaction'])

test['Gender'] = le.fit_transform(test['Gender'])
test['Customer Type'] = le.fit_transform(test['Customer Type'])
test['Type of Travel'] = le.fit_transform(test['Type of Travel'])
test['Class'] = le.fit_transform(test['Class'])
test['satisfaction'] = le.fit_transform(test['satisfaction'])
```

```python
from sklearn.model_selection import train_test_split

y = train['satisfaction']
X = train.drop(columns = ['satisfaction','id'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_test, pred, average='macro')
print(f1_score)
```

```python
test_X = test.drop(columns = ['satisfaction','id'])
pred2 = model.predict(test_X)
#print(pred2)
```

```python
submission = pd.DataFrame()
submission['id'] = test['id']
submission['satisfaction'] = pred2
submission.head()
```

```python
submission.to_csv('flying.csv', index=False)
```
