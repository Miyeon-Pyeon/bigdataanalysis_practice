```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df.head()
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
# df1 df2     

total = pd.concat([df1,df2])
total
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
# df3 df4     .      

pd.concat([df3,df4], join='inner')
```

```python
# df3 df4     .   ,  0 

pd.concat([df3,df4], join='outer').fillna(0)
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
# df5 df6     merge  . Algeria key       

pd.merge(df5, df6, on='Algeria', how='inner')
```

```python
# df5 df6     merge  . Algeria key   

pd.merge(df5, df6, on='Algeria', how='outer')
```
